```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json
from functools import reduce

statesofInterest = ['IOWA', 'INDIANA', 'ILLINOIS', 'KANSAS', 'MINNESOTA', 'MISSOURI', 'NEBRASKA', 'NORTH DAKOTA', 'SOUTH DAKOTA', 'OHIO']
```

```python
hd1 = ["identifier", "jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
pdsi1  = pd.read_csv("weatherData/climdiv-pdsist",  names=hd1, sep='\s+', dtype="str")
#phdi1  = pd.read_csv("weatherData/climdiv-phdidv", names=hd1, sep='\s+', dtype="str")
pmdi1  = pd.read_csv("weatherData/climdiv-pmdist", names=hd1, sep='\s+', dtype="str")
zndx1  = pd.read_csv("weatherData/climdiv-zndx", names=hd1, sep='\s+', dtype="str")
states  = pd.read_csv("weatherData/StatesCode.csv", names=["stateCode" , "State"], dtype="str")
```

```python
def droughtData(states, feature1, state1, month1, featureName):
        st1 = states[states["State"] == state1]["stateCode"].astype(str)
        ident1  = feature1["identifier"].astype(str)
        subset1 = feature1[ident1.str.startswith(st1.iloc[0], na=False)]
        temp1   = subset1[month1]
        year1   = subset1["identifier"].str.strip().str[-4:]    
     
        return(pd.DataFrame({"year": year1.T, "state": state1, featureName + "_" + month1: temp1.T}))                              
```

```python
# Import all states
allStates = states["State"]
      
mnt1 = ["apr", "may", "jun", "jul", "aug", "sep", "oct"]
states1 = allStates
allStates_pdsi1 = []

for j in states1:
    perState = []
    for i in mnt1:
        dt1 = droughtData(states, pdsi1, j, i, "pdsi1")                                           
        perState.append(dt1)
    df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['year'], how='inner', suffixes=(None, "_u")), perState)
    allStates_pdsi1.append(df_merged)
    
    
allStates_pdsi1 = pd.concat(allStates_pdsi1)
allStates_pdsi1.drop(["state_u"], axis = 1, inplace=True)   
allStates_pdsi1.replace('-99.99', np.nan,  inplace=True)
allStates_pdsi1.to_csv("allStates_pdsi1.csv", index = False)
#allStates_pdsi1.head(10)
```
